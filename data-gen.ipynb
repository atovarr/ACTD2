Generación de muestras de una variable aleatoria normal usando numpy.

```python
import numpy as np

mu = 2
sigma = 0.5
n=1000
vals1 = np.random.lognormal(mean=mu, sigma=sigma, size=n)
print(vals1)
```

```python
import numpy as np

mu = 3
sigma = 0.5 
n=1000
vals = np.random.normal(loc=mu, scale=sigma, size=n)
print(vals)
```

Cálculo de estadísticas usando pandas.

```python
import pandas as pd

df1 = pd.DataFrame(vals1)
print(df1.describe())
```

```python
import pandas as pd

df = pd.DataFrame(vals)
print(df.describe())
```

Generación de histograma usando matplotlib.

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(x=vals1, bins=10)
plt.title('Histograma')
plt.xlabel('Tiempos')
plt.ylabel('Frecuencia')
plt.show()

print(bins)
print(count)
```

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(x=vals, bins=30)
plt.title('Histograma de tiempos de servicio')
plt.xlabel('Tiempos de servicio')
plt.ylabel('Frecuencia')
plt.show()

print(bins)
print(count)
```

Densidad normal estándar.

```python
from scipy.stats import lognorm
import numpy as np
import matplotlib.pyplot as plt

x1= np.arange(0,6,0.001)
plt.plot(x1, lognorm.pdf(x1,1))
plt.title('Función de densidad lognormal estándar')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()

from scipy.stats import lognorm
import numpy as np
import matplotlib.pyplot as plt

# Lognormal (mu=3, sigma=1)
x1= np.arange(0,6,0.001)
plt.plot(x1, lognorm.pdf(x1,loc=3,s=1))
plt.title('Función de densidad lognormal(mu=3, sigma=1)')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()

# Lognormal (mu=3, sigma=0.5)
x1= np.arange(0,6,0.001)
plt.plot(x1, lognorm.pdf(x1,loc=3,s=0.5))
plt.title('Función de densidad lognormal(mu=3, sigma=0.5)')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Normal (mu=0, sigma=1)
x= np.arange(-4,4,0.001)
plt.plot(x, norm.pdf(x))
plt.title('Función de densidad normal estándar')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

Densidad normal con media 3 y desviación estándar 1.

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Normal (mu=3, sigma=1)
x= np.arange(0,6,0.001)
plt.plot(x, norm.pdf(x,loc=3))
plt.title('Función de densidad normal(mu=3, sigma=1)')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

Densidad normal con media 3 y desviación estándar 0.5.

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Normal (mu=3, sigma=0.5)
x= np.arange(0,6,0.001)
plt.plot(x, norm.pdf(x,loc=3,scale=0.5))
plt.title('Función de densidad normal(mu=3, sigma=0.5)')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```
